```python
RELATIONS = ['and', 'suc 1', 'suc 2', 'successor', 'predecessor']
MAJOR = 'Computaci√≥n 2013'
```

```python
# Leer los majors desde el archivo
import json
import pandas as pd
import networkx as nx  
import matplotlib.pyplot as plt

with open('majors.json', 'r', encoding='utf-8') as json_majors:
    j = json_majors.read()
    majors = json.loads(j)
```

```python
# Abrir el archivo json correspondiente al major
def open_relation_json(major):
    with open("data/{} - relations c0.json".format(major)) as json_relations:
        j = json_relations.read()
        return json.loads(j)
```

```python
# Abrir json de relaciones
rel = open_relation_json(MAJOR)
def create_table(course_list, course_id, relations, rel_json):
    table = []
    for r in relations:
        table.append(rel_json[r][course_id])
    return pd.DataFrame(table, index=relations, columns=course_list)

# Guardar todas las tablas en una lista
def all_tables(course_list, relations, rel_json):
    alltables = {}
    for i in range(len(course_list)):
        alltables[course_list[i]] = create_table(course_list, i, relations, rel_json)
    return alltables
```

```python
# Mostrar las tablas de cada curso.
for i in range(len(majors[MAJOR])):
    print(majors[MAJOR][i])
    display(create_table(majors[MAJOR], i, RELATIONS, rel))
#     A = create_table(majors[MAJOR], i, RELATIONS, rel)
#     A.to_csv('data/{}.csv'.format(majors[MAJOR][i]))
```

```python
# Crear tabla de relaciones
alltables = all_tables(majors[MAJOR], RELATIONS, rel)
course_matrix = []
for course in alltables:
    course_matrix.append(alltables[course].idxmax(axis=0).tolist())
    
df = pd.DataFrame(course_matrix, index=majors[MAJOR], columns=majors[MAJOR])
# df.to_csv('data/Comp2013/tabla_relaciones.csv')
display(df)
```
